```python
import numpy as np
import matplotlib.pyplot as plt
# 난수로 N차원 배열 생성하기

# np.random.normal()
arr = np.random.normal(0,1,(2,3)) # (loc, scale, size) (평균, 표준편차, 표본개수)
print(arr)

```

```python
arr = np.random.normal(0,1,100000)
plt.hist(arr, bins=100)
plt.show()
```

```python
# np random.rand() 0 ~ 1 사이 
arr = np.random.rand(1000) # 표본개수
plt.hist(arr, bins=100)
plt.show()
```

```python
# np.random.randn() -1 ~ 1 가우시안 분포 표본 추출 함수 
arr = np.random.randn(1000)
plt.hist(arr, bins=100)
plt.show()
```

```python
# np.random.randint() random한 정수만 출력 
arr = np.random.randint(low=1, high=5, size=(2,5)) # 1이상 5미만 
print(arr)
```

```python
arr = np.random.randint(100,200,1000)
plt.hist(arr, bins=100)
plt.show()
```
